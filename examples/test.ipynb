```python
%%manim -qh -v WARNING ThreeDMoebiusTorus

from manim import *


class ThreeDMoebiusTorus(ThreeDScene):
    def construct(self):
        # 定义扭曲成莫比乌斯环形状的圆环柱体的参数方程
        def moebius_torus_function(u, v):
            R = 2  # 大环半径
            r = 1  # 小环半径
            # 调整参数方程以实现扭曲效果
            x = (R + r * np.cos(v)) * np.cos(u)
            y = (R + r * np.cos(v)) * np.sin(u)
            # 在 z 方向引入扭转
            z = r * np.sin(v) * np.sin(u / 2)
            return np.array([x, y, z])

        # 创建莫比乌斯环形状的圆环柱体表面
        moebius_torus = Surface(
            moebius_torus_function,
            u_range=[0, TAU],
            v_range=[0, TAU],
            checkerboard_colors=[BLUE_D, BLUE_E],
            resolution=(32, 16)
        )

        # 设置相机视角
        self.set_camera_orientation(phi=75 * DEGREES, theta=-90 * DEGREES)
        self.add(moebius_torus)
        # # 添加莫比乌斯环形状的圆环柱体到场景
        # self.play(Create(moebius_torus))
        #
        # # 旋转莫比乌斯环形状的圆环柱体
        # self.begin_ambient_camera_rotation(rate=0.1)
        # self.wait(10)
```

```python
%%manim -qh -v WARNING MatrixDeterminant

class MatrixDeterminant(Scene):
    def construct(self):
        # 定义矩阵
        matrix_data = [
            [-18, 23, 39],
            [42, 96, 87],
            [33, 15, -54]
        ]

        # 创建矩阵对象
        matrix = Matrix(matrix_data)
        matrix_title = Text("3×3矩阵的行列式计算", font_size=24).to_edge(UP)

        # 显示矩阵和标题
        self.play(Write(matrix_title))
        self.play(Create(matrix))
        self.wait(1)

        # 移动矩阵到左侧
        self.play(matrix.animate.to_edge(LEFT))
        self.wait(0.5)

        # 展示行列式公式
        formula = MathTex(
            r"\begin{vmatrix} a & b & c \\ d & e & f \\ g & h & i \end{vmatrix}",
            r"=",
            r"a \cdot \begin{vmatrix} e & f \\ h & i \end{vmatrix}",
            r"- b \cdot \begin{vmatrix} d & f \\ g & i \end{vmatrix}",
            r"+ c \cdot \begin{vmatrix} d & e \\ g & h \end{vmatrix}"
        ).scale(0.8).to_edge(RIGHT)

        self.play(Write(formula))
        self.wait(2)

        # 替换为实际数值
        actual_formula = MathTex(
            r"\begin{vmatrix} -18 & 23 & 39 \\ 42 & 96 & 87 \\ 33 & 15 & -54 \end{vmatrix}",
            r"=",
            r"-18 \cdot \begin{vmatrix} 96 & 87 \\ 15 & -54 \end{vmatrix}",
            r"- 23 \cdot \begin{vmatrix} 42 & 87 \\ 33 & -54 \end{vmatrix}",
            r"+ 39 \cdot \begin{vmatrix} 42 & 96 \\ 33 & 15 \end{vmatrix}"
        ).scale(0.8).to_edge(RIGHT)

        self.play(Transform(formula, actual_formula))
        self.wait(2)

        # 计算第一个2x2行列式
        det1 = MathTex(
            r"\begin{vmatrix} 96 & 87 \\ 15 & -54 \end{vmatrix}",
            r"=",
            r"96 \cdot (-54) - 87 \cdot 15",
            r"=",
            r"-5184 - 1305",
            r"=",
            r"-6489"
        ).scale(0.7).next_to(matrix, DOWN, buff=1).shift(RIGHT*2)

        self.play(Write(det1[:3]))
        self.wait(1)
        self.play(Write(det1[3:5]))
        self.wait(1)
        self.play(Write(det1[5:]))
        self.wait(2)

        # 计算第二个2x2行列式
        det2 = MathTex(
            r"\begin{vmatrix} 42 & 87 \\ 33 & -54 \end{vmatrix}",
            r"=",
            r"42 \cdot (-54) - 87 \cdot 33",
            r"=",
            r"-2268 - 2871",
            r"=",
            r"-5139"
        ).scale(0.7).next_to(det1, DOWN, buff=0.7)

        self.play(Write(det2[:3]))
        self.wait(1)
        self.play(Write(det2[3:5]))
        self.wait(1)
        self.play(Write(det2[5:]))
        self.wait(2)

        # 计算第三个2x2行列式
        det3 = MathTex(
            r"\begin{vmatrix} 42 & 96 \\ 33 & 15 \end{vmatrix}",
            r"=",
            r"42 \cdot 15 - 96 \cdot 33",
            r"=",
            r"630 - 3168",
            r"=",
            r"-2538"
        ).scale(0.7).next_to(det2, DOWN, buff=0.7)

        self.play(Write(det3[:3]))
        self.wait(1)
        self.play(Write(det3[3:5]))
        self.wait(1)
        self.play(Write(det3[5:]))
        self.wait(2)

        # 汇总结果
        final_calc = MathTex(
            r"=& -18 \cdot (-6489) - 23 \cdot (-5139) + 39 \cdot (-2538) \\",
            r"=& 116802 + 118197 - 98982 \\",
            r"=& 136017"
        ).scale(0.8).next_to(det3, DOWN, buff=0.7).shift(LEFT*0.5)

        self.play(Write(final_calc[0]))
        self.wait(1)
        self.play(Write(final_calc[1]))
        self.wait(1)
        self.play(Write(final_calc[2]))
        self.wait(2)

        # 最终结果框
        result_box = SurroundingRectangle(final_calc[2], buff=0.2, color=YELLOW)
        result_text = Text("行列式值 = 136017", font_size=24).next_to(result_box, DOWN)

        self.play(Create(result_box))
        self.play(Write(result_text))
        self.wait(3)
```

```python
class Book(VGroup):
    def __init__(
            self,
            title,
            index=0,
            fill_color=GRAY,
            stroke_color=GRAY_E
    ):
        super().__init__()
        self.title = title
        self.index = index

        t = VGroup()
        for i in self.title:
            t.add(Text(i))
        t.arrange(DOWN, buff=0.1)

        depth_points = [
            np.array((0.25, -1, 0.0)),
            np.array((0.6, -0.25, 0.0)),
            np.array((0.6, -0.25, 0.0))+ 2*UP,
            np.array((0.25, -1, 0.0)) + 2*UP,
        ]

        # 创建平行四边形
        depth_pol = Polygon(
            *depth_points,
            fill_color=fill_color,
            stroke_color=stroke_color,
            fill_opacity=1,
            stroke_opacity=1,
            stroke_width=2
        )

        height_points = [
            np.array((-0.25, 1, 0.0)),
            np.array((0.1, 1.75, 0.0)),
            np.array((0.1, 1.75, 0.0))+ 0.5*RIGHT,
            np.array((-0.25, 1, 0.0)) + 0.5*RIGHT,
        ]

        height_pol = Polygon(
            *height_points,
            fill_color="#ffeebb",
            stroke_color=stroke_color,
            fill_opacity=1,
            stroke_opacity=1,
            stroke_width=2
        )

        self.add(
            depth_pol,
            height_pol,
            Rectangle(
                fill_color=fill_color,
                stroke_color=stroke_color,
                fill_opacity=1,
                stroke_opacity=1,
                height=2,
                width=0.5,
            ),
            t.shift(DOWN*0.4).scale(0.8),

        )
        self.set_z_index(self.index)
```
