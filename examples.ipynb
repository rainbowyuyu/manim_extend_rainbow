```python
%%manim -qh -v WARNING SquTexIndex

color_list = [RED,ORANGE,YELLOW,GREEN,BLUE,TEAL,PURPLE]

class SquTexIndex(Scene):
    def construct(self):
        s = SquTex("rainbow").shift(UP*1.5)
        for i in range(len(s)):
            s[i].set_color(color_list[i])
            
        s_cp = s.copy().shift(DOWN*3)
        s_cp[0][0].set_opacity(0)
        s_cp[1].set_color(WHITE)
        s_cp[2].shift(DOWN*2)
        s_cp.change_word_in_text(3,"*")
        s_cp[4].scale(0.5).rotate(np.pi/4)
        s_cp.change_text(5,color = WHITE)
        s_cp.change_square(6,color = WHITE)
        self.add(s,s_cp)
```

```python
%%manim -qh -v WARNING SquTexCreate
class SquTexCreate(Scene):
    def construct(self):
        t = SquTex("rainbow")
        self.play(t.animate_one_by_one(FadeIn , scale=1.5))
        self.wait()
```

```python
%%manim -qh -v WARNING TestFade
class TestFade(Scene):
    def construct(self):
        a = SquTexSlide("ytyabc")
        b = SquTex("aa")
        self.add(a)
        self.play(*a.slide_fade('forward',b))
        self.play(*a.slide_fade(2))
```

```python
%%manim -qh -v WARNING MatrixExample

color_by_gradient_1 = [
    RED,
    ORANGE,
    YELLOW,
]
color_by_gradient_2 = [
    BLUE,
    TEAL,
    GREEN,
]

class MatrixExample(Scene):
    def construct(self):
        m1 = MatrixCal(matrix_3)
        m2_o = MatrixCal(matrix_3,brackets_pair=['(',')'])
        m2 = m2_o.neg_with_brackets()
        m3 = MatrixCal(matrix_2t3)
        m3_c = m3.get_column(0)
        m3_c.set_color_by_gradient(color_by_gradient_1)
        m4 = MatrixCal(matrix_3t2)
        m4_r = m4.get_row(1)
        m4_r.set_color_by_gradient(color_by_gradient_2)
        sbg = Square(fill_opacity=0.5,stroke_opacity=0).set_color(BLUE)
        sbg.scale(0.6)
        m2[0][0].shift(UP*0.5)
        m2[0][1].set_color(RED)
        m2[0][2].scale(1.5).set_color(GREEN)
        m2[2][2].rotate(PI/2)
        m_v_1 = VGroup(m1,m2).arrange(buff=1)
        m_v_2 = VGroup(m3,m4).arrange(buff=1)
        m_v = VGroup(m_v_1,m_v_2).arrange(buff=0.5,direction=DOWN)
        sur = SurroundingRectangle(m2[2][0],buff=0)
        sbg.move_to(m2[2][1])
        self.add(m_v,sur,sbg)
```
